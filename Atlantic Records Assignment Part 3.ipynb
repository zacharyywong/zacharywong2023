```python
# Extract the Top 5 Most Important Songs and the cities that it has broken into the Top20 Charts 
import pandas as pd

df = pd.read_csv('/Users/zacharywong/github/zacharywong2023/MostImportantSongsinUS.csv')
df
```

```python
#Set index to Song name and artist
df = df.set_index(['Song Name', 'Artist'])
df_new = df.loc[:, 'New York City':'Orlando']
df_new
```

```python
# Create a dictionary out of the dataframe to extract all cities that each song is in its Top20 Chart
# Rotate columns and axis to make the songs/artists keys in the dictionary 
df_new = df_new.T

# Total dataframe in dictionary form 
# keys = song/artist tuple 
# Value = dictionary where key = column name and value = element from dataframe
dict = df_new.to_dict()
dict
```

```python
#Populate a new dictionary where its keys = songs/artist and values = all cities that the song is in its top20 chart

SongCitiesDict  = {}

# Extract song/artist name 
for x in dict:
    CityList = []
    CityDict = dict.get(x)
    for y in CityDict:
        if pd.isna(CityDict.get(y)) == False:
            CityList.append(y)
    SongCitiesDict[x] = CityList

SongCitiesDict['You Problem', 'MONSTA X']
```
